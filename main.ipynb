```python
df = pd.read_excel('MDC1_BL.xlsx')
activity_id_map = {id_: i + 1 for i, id_ in enumerate(df['Activity ID'])}

df['Activity ID'] = df['Activity ID'].map(activity_id_map)

import re

def replace_ids_in_dependency_column(dep_col):
    def parse_and_replace(cell):
        if pd.isna(cell):
            return []
        items = [item.strip() for item in cell.split(',')]
        new_items = []
        for item in items:
            match = re.match(r'([A-Za-z0-9.\-]+)(:\s*\w+\s*-?\d*)?', item)
            if match:
                original_id = match.group(1)
                suffix = match.group(2) or ''
                if original_id in activity_id_map:
                    new_items.append(str(activity_id_map[original_id]) + suffix)
                else:
                    new_items.append(item)
        return new_items
    return dep_col.apply(parse_and_replace)

df['Predecessor Details'] = replace_ids_in_dependency_column(df['(*)Predecessor Details'])
df['Successor Details'] = replace_ids_in_dependency_column(df['(*)Successor Details'])

# import ace_tools as tools; tools.display_dataframe_to_user(name="Simplified Tasks Data", dataframe=df)
```

```python
from datetime import timedelta

in_progress_df = df[df['Activity Status'] == 'In Progress'].copy()
in_progress_df['% Complete'] = in_progress_df['Activity % Complete(%)'] / 100
in_progress_df['Expected Duration'] = (in_progress_df['Actual Duration'] / in_progress_df['% Complete'].replace(0, 0.1)) * (1 - in_progress_df['% Complete'])
in_progress_df['Expected Duration'] += in_progress_df['Actual Duration']
in_progress_df['Estimated End Date'] = in_progress_df['Actual Start'] + in_progress_df['Expected Duration'].apply(lambda x: timedelta(days=x))

# Map estimated end and start dates
estimated_end_map = dict(zip(in_progress_df['Activity ID'], in_progress_df['Estimated End Date']))
actual_start_map = dict(zip(df['Activity ID'], df['Actual Start']))
actual_finish_map = dict(zip(df['Activity ID'], df['Actual Finish']))

# Prepare Not Started tasks
not_started_df = df[df['Activity Status'] == 'Not Started'].copy()

def parse_pred_string(pred_str):
    if pd.isna(pred_str):
        return []
    return [p.strip() for p in pred_str.split(',')]

def estimate_ns_start(pred_list):
    latest_date = pd.NaT
    for pred in pred_list:
        try:
            if ':' not in pred:
                pred += ": FS"
            pred_id, relation = pred.split(':')
            pred_id = pred_id.strip()
            relation = relation.strip()
            print(pred_id in estimated_end_map)

            # Handle lag (e.g., "SS -1")
            lag = 0
            if '-' in relation:
                parts = relation.split('-')
                relation = parts[0].strip()
                lag = int(parts[1].strip())

            if relation == 'FS':
                ref_date = estimated_end_map.get(pred_id, actual_finish_map.get(pred_id))
            elif relation == 'SS':
                ref_date = actual_start_map.get(pred_id)
            elif relation == 'FF':
                ref_date = actual_finish_map.get(pred_id)
            elif relation == 'SF':
                ref_date = actual_start_map.get(pred_id)
            else:
                ref_date = None

            if pd.notnull(ref_date):
                ref_date = pd.to_datetime(ref_date) + timedelta(days=lag)
                latest_date = max(latest_date, ref_date) if pd.notnull(latest_date) else ref_date

        except Exception:
            continue
    return latest_date

# Apply estimation
not_started_df['Parsed Predecessors'] = not_started_df['Predecessor Details'].apply(parse_pred_string)
not_started_df['Estimated Start Date'] = not_started_df['Parsed Predecessors'].apply(estimate_ns_start)

# Save into main dataframe
df.loc[in_progress_df.index, 'Estimated End Date'] = in_progress_df['Estimated End Date']
df.loc[not_started_df.index, 'Estimated Start Date'] = not_started_df['Estimated Start Date']
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = df[df['Activity Status'].isin(['Not Started', 'In Progress'])].copy()
df['Parsed Successors'] = df['Successor Details'].fillna("").apply(
    lambda x: [s.split(':')[0].strip() for s in x.split(',') if ':' in s]
)
dependency_map = {row['Activity ID']: row['Parsed Successors'] for _, row in df.iterrows()}

top_n_tasks = sorted(dependency_map.items(), key=lambda x: len(x[1]), reverse=True)[:15]
top_task_ids = [task for task, _ in top_n_tasks]

def trace_downstream(task, visited):
    if task in visited:
        return set()
    visited.add(task)
    affected = set()
    for succ in dependency_map.get(task, []):
        affected.add(succ)
        affected |= trace_downstream(succ, visited)
    return affected

cascade_count = {}
for task in top_task_ids:
    affected = trace_downstream(task, set())
    cascade_count[task] = len(affected)

cascade_series = pd.Series(cascade_count).sort_values(ascending=False)
plt.figure(figsize=(10, 5))
cascade_series.plot(kind='bar', color='#2196F3')
plt.title("Top 15 Tasks by Downstream Dependencies")
plt.ylabel("Number of Affected Tasks")
plt.xlabel("Root Task")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import ast
from collections import defaultdict
```

```python
df = pd.read_excel("MDC1_BL_1.xlsx")
df = df.rename(columns={'(*)Activity Type': 'Activity Type', '(*)Planned Start': 'Planned Start', '(*)Planned Finish': 'Planned Finish', '(*)Actual Duration(d)': 'Actual Duration', 'Remaining Duration(d)': 'Remaining Duration', '(*)Total Float(d)': 'Total Float', '(*)Remaining Float(d)': 'Remaining Float', 'Original Duration(d)': 'Original Duration'})
```

```python
in_progress_ids = df[df['Activity Status'] == 'In Progress']['Activity ID'].tolist()
not_started_df = df[df['Activity Status'] == 'Not Started'].copy()
not_started_df['Parsed Predecessors'] = not_started_df['Predecessor Details'].apply(ast.literal_eval)
# st.dataframe(not_started_df)

dependency_count = defaultdict(int)
for _, row in not_started_df.iterrows():
    # st.header(row)
    preds = [int(p.split(':')[0]) for p in row['Parsed Predecessors']]
    for pid in preds:
        if pid in in_progress_ids:
            dependency_count[pid] += 1

dependency_df = pd.DataFrame(dependency_count.items(), columns=['In Progress Task ID', '# Not Started Dependents'])
dependency_df = dependency_df.sort_values(by='# Not Started Dependents', ascending=False)
```

```python
in_progress_df = df[df['Activity Status'] == 'In Progress'].copy()
in_progress_df['% Complete'] = in_progress_df['Activity % Complete(%)'] / 100
in_progress_df['Estimated Duration'] = (in_progress_df['Actual Duration'] / in_progress_df['% Complete'].replace(0, 0.1)) * (1 - in_progress_df['% Complete'])
in_progress_df['Estimated Duration'] = in_progress_df["Estimated Duration"] + in_progress_df["Actual Duration"]
in_progress_df['Max Duration Inc. Float'] = in_progress_df['Original Duration'] + in_progress_df['Total Float']

def classify(row):
    if row['Estimated Duration'] <= row['Original Duration']:
        return 'On-Time'
    elif row['Estimated Duration'] <= row['Max Duration Inc. Float']:
        return 'Can be on-time using float'
    else:
        return 'Delayed'

in_progress_df['Status'] = in_progress_df.apply(classify, axis=1)
status_counts = in_progress_df['Status'].value_counts()
```
